```python
ds = pd.read_csv('/kaggle/input/heart-failure-prediction/heart.csv')
ds.head()
```

```python
n = pd.get_dummies(ds,columns=['Sex','ChestPainType','RestingECG','ExerciseAngina','ST_Slope'])
n
```

```python
# rows and columns in the data set
n.shape
```

```python
n.info()
```

```python
n.isnull().sum()
#n.tail()
```

```python
n['HeartDisease'].value_counts()
```

```python
Y = n['HeartDisease']
X = n.drop(columns='HeartDisease',axis=1)
X
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=1)

X_train.shape
y_train.shape
X_test.shape
y_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(X_train,y_train)
```

```python
# MODEL ACCURACY

from sklearn.metrics import accuracy_score

# prediction from training data
X_train_prediction = lr.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction,y_train)
print("accuracy on test data : ",training_data_accuracy)
```

```python
# testing data accuracy

X_test_prediction = lr.predict(X_test)
testing_data_accuracy = accuracy_score(X_test_prediction,y_test)
print("accuracy on test data : ",testing_data_accuracy)
```
